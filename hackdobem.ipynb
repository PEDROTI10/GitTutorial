```python
#Entrada de dados

# n >= 9 == A
# 9 > n >= 7 == B
# 7 > n >= 6 == C
# 6 > n >= 5 == D
# N < 5 == E

nota = float(input("Digite a nota do aluno: "))

#Processamento e saída dos dados

if (nota >= 9):
  print("Conceito A")
elif (nota >= 8):
  print("Conceito B")
elif (nota >= 7):
  print("Conceito C")
elif (nota >= 6):
  print("Conceito D")
else:
  print(f"Reprovado: Conceito E")
print("Fim do programa")
```

```python
# Recebendo as idades dos usuarios

id1 = input("Digite a idade do primeiro usuario: \n")
id2 = input("Digite a idade do segundo usuario: \n")
id3 = input("Digite a idade do terceiro usuario: \n")
id4 = input("Digite a idade do quarto usuario: \n")
id5 = input("Digite a idade do quinto usuario: \n")

# Convertando um numero para inteiro

id1 = int(id1)
id2 = int(id2)
id3 = int(id3)
id4 = int(id4)
id5 = int(id5)

# Calculando a media

media = (id1+id2+id3+id4+id5)/5

# Mostrando a media das idades

print("MEDIA:", media)
```

```python
# Criando a variavel de controle do laco
contador = 1

# Criando a variavel para acumular os valores das idades digitadas
soma = 0

#Criando o laco de repeticao para receber 5 idades e soma-las
while (contador <= 5):

# Acumula as idaades dos usuarios

  idade = int(input(f"Digite a idade do nº{contador} "))

# Somando a idade do usuario ao valor atual da variavel acumulador
  soma = soma + idade

# Incrementa o contador
  contador = contador + 1

# Calculando a media das idades
media = soma/(contador -1)
print(f"A media das idades digitadas e igual: {media}")


```
