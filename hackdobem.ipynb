```python
#Entrada de dados

# n >= 9 == A
# 9 > n >= 7 == B
# 7 > n >= 6 == C
# 6 > n >= 5 == D
# N < 5 == E

nota = float(input("Digite a nota do aluno: "))

#Processamento e saída dos dados

if (nota >= 9):
  print("Conceito A")
elif (nota >= 8):
  print("Conceito B")
elif (nota >= 7):
  print("Conceito C")
elif (nota >= 6):
  print("Conceito D")
else:
  print(f"Reprovado: Conceito E")
print("Fim do programa")
```
